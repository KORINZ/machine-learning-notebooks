## Code Example 1-1 Better Life Index (BLI) 2020 Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Default plot settings
plt.rc("font", size=14)
plt.rc("axes", titlesize=14, labelsize=14)
plt.rc("legend", fontsize=12)
plt.rc("xtick", labelsize=10)
plt.rc("ytick", labelsize=10)
```

```python
# Download the data
data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
X = lifesat[['GDP per capita (USD)']].values
y = lifesat[['Life satisfaction']].values

# Visualize the data
lifesat.plot(kind="scatter", grid=True, x="GDP per capita (USD)", y="Life satisfaction")
plt.show()

# Select a linear model; model-based learning
model = LinearRegression()

# Train the model
model.fit(X, y)

# Make a prediction for Japan
X_new = np.array([[39918]])  # Japan's GDP per capita (USD) in 2020
print(model.predict(X_new)) # outputs [[6.45505061]]

# Print the coefficient and intercept
print(model.coef_) # outputs [[6.77889969e-05]]
print(model.intercept_) # outputs [3.74904943]
```

```python
# Try k-Nearest Neighbors Regression model; instance-based learning
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=3)

# Train the model
model.fit(X, y)

# Make a prediction
print(model.predict(X_new)) # outputs [[6.66666667]]
```
